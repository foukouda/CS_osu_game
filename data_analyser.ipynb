```python
file_path = './game_results.csv'
game_data = pd.read_csv(file_path, encoding='latin1')

# Aperçu des données
print("Aperçu des données:")
print(game_data.head())

# Informations générales
print("\nInformations générales:")
print(game_data.info())

# Description statistique des données
print("\nDescription statistique des données:")
print(game_data.describe())

# Analyser la répartition des Hits
if 'Hits' in game_data.columns:
    plt.figure(figsize=(10, 6))
    sns.histplot(game_data['Hits'], bins=20, kde=True)
    plt.title('Distribution des Hits')
    plt.xlabel('Hits')
    plt.ylabel('Fréquence')
    plt.show()
else:
    print("La colonne 'Hits' n'existe pas dans les données.")

# Comparer les Misses par joueur
if 'Nom du Joueur' in game_data.columns and 'Misses' in game_data.columns:
    plt.figure(figsize=(14, 8))
    sns.barplot(data=game_data, x='Nom du Joueur', y='Misses', estimator=sum)
    plt.title('Nombre de Misses par Joueur')
    plt.xlabel('Joueur')
    plt.ylabel('Total Misses')
    plt.xticks(rotation=45)
    plt.show()
else:
    print("Les colonnes 'Nom du Joueur' ou 'Misses' n'existent pas dans les données.")

# Analyser la corrélation entre les variables numériques
numeric_columns = game_data.select_dtypes(include=['number']).columns
if len(numeric_columns) > 1:
    plt.figure(figsize=(10, 6))
    sns.heatmap(game_data[numeric_columns].corr(), annot=True, cmap='coolwarm')
    plt.title('Matrice de corrélation')
    plt.show()
else:
    print("Il n'y a pas suffisamment de colonnes numériques pour effectuer une analyse de corrélation.")

# Afficher les performances moyennes par joueur (Précision Moyenne)
if 'Nom du Joueur' in game_data.columns and 'Précision Moyenne' in game_data.columns:
    player_avg_precision = game_data.groupby('Nom du Joueur')['Précision Moyenne'].mean().reset_index()
    print("\nPrécision Moyenne par joueur:")
    print(player_avg_precision)

    # Affichage des précisions moyennes
    plt.figure(figsize=(14, 8))
    sns.barplot(data=player_avg_precision, x='Nom du Joueur', y='Précision Moyenne')
    plt.title('Précision Moyenne par Joueur')
    plt.xlabel('Joueur')
    plt.ylabel('Précision Moyenne')
    plt.xticks(rotation=45)
    plt.show()
else:
    print("Les colonnes 'Nom du Joueur' ou 'Précision Moyenne' n'existent pas dans les données.")
```
